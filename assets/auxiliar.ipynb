```python
def last_bid_ask(*symbols):
    last = {}
    bid = {}
    ask = {}
    for symbol in symbols:
        last[symbol] = mt5.symbol_info_tick(symbol).last
        bid[symbol] = mt5.symbol_info_tick(symbol).bid
        ask[symbol] = mt5.symbol_info_tick(symbol).ask
    return last, bid, ask

pymt5.conecta_mt5("BTG")
last, bid, ask = pymt5.last_bid_ask(*symbols)
```
